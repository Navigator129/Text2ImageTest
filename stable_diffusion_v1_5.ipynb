```python
def fetch_prompt(path):
  prompts = []
  with open(path, 'r', encoding = 'utf-8') as f:
      dict_list = json.load(f)

  for dict_ in dict_list:
      prompts.append(dict_['prompt'])
  return prompts
```

```python
from diffusers import StableDiffusionPipeline
import torch
from tqdm import tqdm
import os

model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
quick_test_path1 = '/content/drive/MyDrive/Prompts/test.json'
prompts = fetch_prompt(quick_test_path1)
len(prompts)
```

```python
idx = 0
for p in tqdm(prompts):
  save_path = "/content/drive/MyDrive/New_exp/sd1_5/quick_test/"
  img_path = os.path.join(save_path, "{}.png".format(idx))
  if os.path.exists(img_path):
    continue
  image = pipe(p).images[0]
  # Save image
  image.save(img_path)
  idx += 1
```
